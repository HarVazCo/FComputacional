**(c)** Modifica nuevamente tu programa para calcular y graficar la *media (promedio) móvil* de los datos, definida por:
$$ Y_k = \dfrac{1}{2r} \sum_{m=-r}^{r} y_{k+m},$$
donde $r=5$ (en este caso) y $y_k$ son los números de manchas solares.

El programa debe graficar tanto los datos originales como la *media móvil* en el mismo gráfico, sólo sobre los primeros 1000 datos.

```python
# Definimos nuestra r y creamos una lista Y1 con la media movil
r = 5 
Y1 = []
X1 = x1[r:len(y1)-r]

for i in range(r,len(y1)-r):
    S = 0
    for j in range(-r,r+1):
        S+= y1[i+j]

    Y1.append((1/(2*r))*S)

# Graficamos los 1000 datos experimentales y la media movil

plt.figure(figsize=(7,6))
plt.axhline(y = 0.0, xmin = 0.0, color = 'black')
plt.axvline(x = 0.0, ymin = 0.0, color = 'black')
plt.plot(x1,y1,label='Datos experimentales')
plt.plot(X1,Y1,color='r',label='Media móvil')
plt.xlabel("t [meses]")
plt.ylabel("y [manchas]")
plt.legend()
plt.grid()
plt.show()
```

Aunque la función `plot` está diseñada principalmente para hacer gráficos $xy$ estándar, se puede adaptar para

**(a)** Grafica la llamada curva *deltoide*, definida paraméticamente por las ecuaciones:
$$x = 2 \cos \theta + \cos 2\theta , \hspace{1cm} y = 2\sin \theta - \sin 2\theta,$$
donde $0\leq \theta \leq 2\pi$. Toma un conjunto de valores para $\theta$ entre $0$ y $2\pi$ y calcula $x(\theta)$ 
e $y(\theta)$ usando las ecuaciones anteriores, para porteriormente graficar $y$ como función de $x$.

```python
# Definimos el dominio 
theta = np.linspace(0,2*np.pi,500)

# Obtenemos el conjunto de puntos x, y 

x = 2* np.cos(theta) + np.cos(2*theta)
y = 2* np.sin(theta) - np.sin(2*theta)

# Graficamos

plt.figure(figsize=(7,7))
plt.axhline(y = 0.0, xmin = 0.0, color = 'black')
plt.axvline(x = 0.0, ymin = 0.0, color = 'black')
plt.plot(x,y,color='m')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

**(b)** Usando este mismo enfoque, se puede hacer una gráfica polar $r=f(\theta)$ para alguna función $f$ calculando $r$ para un rango de valores 
de $\theta$ y luego convirtiendo $r$ y $\theta$ a coordenadas cartesianas usando las ecuaciones estándar:
$$x = r \cos \theta, \hspace{1cm} y = r \sin \theta.$$
Utiliza este método para trazar la *espiral Galileana* $r = \theta ^2$ para $0 \leq \theta \leq 10 \pi$.

```python
# Definimos el dominio 
theta = np.linspace(0,10*np.pi,1000)


# Obtenemos el conjunto de puntos r, x, y

r = theta **2
x = r* np.cos(theta) 
y = r* np.sin(theta) 

# Graficamos

plt.figure(figsize=(7,7))
plt.axhline(y = 0.0, xmin = 0.0, color = 'black')
plt.axvline(x = 0.0, ymin = 0.0, color = 'black')
plt.plot(x,y,color='g')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

## 3. La gráfica de Feigenbaum (caos determinista)


```python
# Definimos el mapeo logistico
def map(x,r):
    return r*x*(1-x)

# Definimos xn, el dominio de r, la dr y el numero de iteraciones
xn = 1/2
dr = 0.001
r = np.arange(1,4+dr,dr)
N = 1000

# Obtenemos los puntos que vamos a graficar, es decir, los puntos (r,xinf)
Xinf = []
R = []

for i in r:
    for j in range(N):
        xinf = map(xn,i)
        xn = xinf
        if j > N-200:
            Xinf.append(xinf)
            R.append(i)
        else:
            continue
```

```python
# Graficamos los puntos obtenidos
plt.figure(figsize=(8,6))
plt.scatter(R,Xinf,marker = '.',s = 0.02,c = 'darkred')
plt.grid()
plt.xlabel("r")
plt.ylabel("x")
plt.show()
```

Notamos que en $r=3$ el comportamiento cambia, en un inicio se tiene un valor específico para $x$, sin embargo, depués del punto llamado *"borde del caos"*, $x$ tiene múltiples valores, en otras palabras, hay un comportamiento caótico.

## 4. El conjunto de Mandelbrot


**(a)** **Escribe un programa para crear una imagen del conjunto de Mandelbrot**

```python
# Damos la instruccion de ignorar errores de numpy, para que no se detenga el programa
np.seterr(all="ignore")
# Damos la N para la cuadricula y hacemos la cuadricula 

N = 8000
x = np.linspace(-2,2,N)
y = np.linspace(-2,2,N)
X,Y = np.meshgrid(x,y)

def c_matriz(x,y):
    return x[np.newaxis, :] + y[:, np.newaxis] * 1j

can = c_matriz(x,y)

# Definimos una funcion que nos de si c esta en el conjunto o no 

def esta(c,n):
    z = 0 
    for i in range(n+1):
        z = z**2 + c
    return abs(z) <= 2

zz = esta(can,100)
```

```python
# Graficamos los puntos
plt.figure(figsize=(8,8))
plt.imshow(zz,cmap='binary',extent=[-2,2,-2,2])
plt.show()
```

## 5. Factorial (valores flotantes)

```python
# Definimos la funcion factorial que nos devuelve enteros
def fact(n):
    f = 1
    for i in range(1, n+1):
       f = f * i
    return f
# Definimos la funcion factorial que devuelve valores flotantes
def fact_float(n):
    f = float(1)
    for i in range(1, n+1):
       f = f * i
    return f
```

```python
fact(200)
```

```python
fact_float(200)
```

Notamos que cuando definimos el factorial de manera que python trabaje con puros números enteros obtenemos un número muy grande, pero concreto, cosa que no pasa cuando hacemos que python trabaje con números flotantes, debido a que es muy grande el flotante le cuesta mucho trabajo operar con este, por lo tanto al final se obtiene como resultado que el número es infinito, algo que no es concreto.

**(b)** Existe otra forma de escribir las soluciones de una ecuación cuadrática.

Demuestra que multiplicando la parte superior e inferior de la solución anterior por $-b\mp \sqrt{b^2-4ac}$, las soluciones también se pueden escribir como:
$$x_{1,2} = \dfrac{2c}{-b\mp \sqrt{b^2-4ac}}.$$
Agrega unas lineas a tu programa para imprimir también estos valores, además de los anteriores y entonces usa nuevamente el programa para resolver $0.001x^2+1000x+0.001=0$. **¿Qué es lo que ves? ¿Cómo lo explicas?**.

**Dem.**

Dado que las soluciones estan dadas por $x_{1,2} = \dfrac{-b \pm \sqrt{b^2-4ac}}{2a}$, multiplicando por $ \dfrac{-b \mp \sqrt{b^2-4ac}}{-b \mp \sqrt{b^2-4ac}}$ del lado derecho se tiene que:

$$ x_{1,2}= \left(\dfrac{-b \pm \sqrt{b^2-4ac}}{2a}\right) \left( \dfrac{-b \mp \sqrt{b^2-4ac}}{-b \mp \sqrt{b^2-4ac}}\right) = \dfrac{b^2 \pm b\sqrt{b^2-4ac} \mp b\sqrt{b^2-4ac} - (b^2 -4ac)}{2a(-b \mp \sqrt{b^2-4ac})}$$
$$ = \dfrac{4ac}{2a(-b \mp \sqrt{b^2-4ac})} = \dfrac{2c}{-b \mp \sqrt{b^2-4ac}}$$
$$ \therefore x_{1,2}=\dfrac{2c}{-b \mp \sqrt{b^2-4ac}}$$

```python
# Le pedimos al usuario ingresar su ecuacion cuadratica
a = float(input('Para una ecuación de la forma ax²+bx+c=0\n a es igual a:'))
b = float(input('b es igual a:'))
c = float(input('c es igual a:'))

# Calculamos las raices con dos metodos distintos
x1 = (-b + sqrt(b**2 - 4*a*c))/(2*a)
x2 = (-b - sqrt(b**2 - 4*a*c))/(2*a)

x11 = (2*c)/(-b - sqrt(b**2 - 4*a*c))
x22 = (2*c)/(-b + sqrt(b**2 - 4*a*c))

print('Las soluciones con el primer método son: x1 =',x1,', x2 =',x2)
print('Las soluciones con el segundo método son: x1 =',x11,', x2 =',x22)
```

Las raices obtenidas son cercanas, pero no idénticas, otra cosa que notamos es que tienen bastantes números después del punto decimal, debido que a python le cuesta más trabajo operar con números después del punto decimal, puede que eso haya propagado el error y al final provocar que los resultados númericos sean distintos.

## 7. Cálculo de derivadas 
Supongamos que tenemos una función $f(x)$ y queremos calcular su derivada en un punto $x_0$. Esto lo podemos hacer con lápiz y papel si conocemos la fórmula matemática de la función, o podemos hacerlo en la computadora haciendo uso de la definición de la derivada: 
$$ \dfrac{df}{dx} = \lim_{\delta x \to 0} \dfrac{f(x_0 +\delta x)-f(x_0)}{\delta x}.$$
En la computadora no podemos tomar el límite cuando $\delta x$ llega a cero, pero podemos obtener una aproximación razonable simplemente haciendo que $\delta x$ sea pequeño.

**(a)** Escribe un programa que defina una función `f(x)` y que devuelva el valor $x(x-1)$, luego calcula la derivada de la función en el punto $x_0=1$ usando la fórmula anterior con $\delta x = 10^{-2}$. Calcula analíticamente el valor real de la misma derivada y compáralo con la respuesta de tu programa. Ambos no estarán perfectamente de acuerdo. **¿Por qué no?**

```python
# Definimos la funcion que nos pide el problema
def f(x):
    return x*(x-1)
# Definimos la funcion derivada
def derf(f,x,dx):
    return (f(x + dx) - f(x))/dx
# Calculamos la derivada de f
derivada = derf(f,1,1e-2)

derivada
```

El valor analítico es 1, los valores no coinciden del todo debido a que teóricamente $\delta x$ debería ser más chico de lo que lo estamos considerando.

**(b)** Repite el cálculo para $\delta = 10^{-4},10^{-6},10^{-8},10^{-10},10^{-12},10^{-14}$. Deberías ver que la precisión del cálculo inicialmente mejora a medida que $\delta x$ se hace más pequeña, pero luego vuelve a empeorar. **¿Por qué pasa esto?**

```python
# Evaluamos la funcion derivada para diferentes valores de dx
dx = [1e-4, 1e-6, 1e-8, 1e-10, 1e-12, 1e-14]
for i in dx:
    print(derf(f,1,i))
```

Aunque teóricamente los valores deberían aproximarse cada vez más a 1, esto no sucede ya que python tiene sus limitaciones al trabajar con números pequeños, hasta cierto punto la aproximación mejora, pero al pasar los limites de python, el error de calculo se propaga y entonces la aproximación deja de mejorar.
